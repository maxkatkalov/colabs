```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
fig, axs = plt.subplots(2, 1)
fig.set_layout_engine("constrained")

n = len(tips["total_bill"]) 
num_bins_sturges = int(1 + np.log2(n))  # https://www.vedantu.com/question-answer/the-sturges-rule-for-determining-the-number-of-class-11-maths-cbse-5fb494704ad6c23c32da71f2

bills_ax = axs[0]
tip_ax = axs[1]

# visualize the distribution of total bills
bills_ax.set_title("Distribution of Total Bills and Tips")
bills_ax.set_xlabel("Total Bill ($)")
bills_ax.set_ylabel("Frequency")

bills_x_ticks = np.arange(0, tips["total_bill"].max() + 2, 2)
bills_ax.set_xticks(bills_x_ticks)

bills_ax.hist(tips["total_bill"], bins=num_bins_sturges)

# visualize the distribution of tips
tip_ax.set_xlabel("Tip amount ($)")
tip_ax.set_ylabel("Frequency")

tip_x_ticks = np.arange(0, tips["tip"].max() + 0.1, 0.5)
tip_ax.set_xticks(tip_x_ticks)

tip_ax.hist(tips["tip"], bins=num_bins_sturges)
```

### Analysis of Tips in dataset:

1. **Range of Tips**: 
   - The minimum tip amount observed is in the **$1–2** range, which shows that the lowest tipping amount is relatively small.
   - The maximum tip amount is in the **$9–10** range, indicating that the largest tips are relatively high compared to the minimum.

2. **Distribution of Tips**:
   - The **highest frequency of tips** occurs in the **$1–2** range, suggesting that most customers tend to leave smaller tips.
   - As the tip amount increases, the frequency decreases. This indicates a **skewed distribution**, where smaller tips are much more common than larger ones.
   
3. **Lack of Tips in the $7.75–$8.85 Range**:
   - Interestingly, there is a noticeable **gap** in the distribution, where no tips fall within the range of **$7.75–$8.85**. This could indicate that people either round their tips to the nearest dollar or there are few cases where the tip falls into this specific range.

4. **Pattern in Larger Tips**:
   - As we move into the **$9–10** range, the frequency of tips continues to be low. This suggests that while larger tips exist, they are less common, and the highest tips are still relatively rare compared to smaller tips.

### Conclusion:
- The tip distribution is **right-skewed**, with the majority of the tips concentrated in the **lower range ($1–2)**, and fewer people leaving larger tips, with a noticeable gap in the **$7.75–$8.85** range.
- This type of distribution is common in service industries where tipping is voluntary, and most customers tend to leave smaller tips, while only a few leave significantly larger ones.

This analysis provides insight into typical tipping behavior, with most customers opting for smaller amounts, and fewer individuals leaving larger tips.
